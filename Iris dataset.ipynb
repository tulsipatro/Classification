# Model Building

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn import metrics

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.decomposition import PCA
```

```python
# Seperating Target and Features

X = data.drop(['species'], axis=1)
y = data['species'].map({
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
})
```

```python
# Splitting the datase
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

# Logistic Regression

```python
# Logistic Regression

model_LR=LogisticRegression()

# Run model on Train data
model_LR.fit(X_train, y_train)

# Predictions
predictions_train = model_LR.predict(X_train)
predictions_test = model_LR.predict(X_test)

# Accuracy
from sklearn.metrics import accuracy_score
print("Accuracy_train : "+str(accuracy_score(y_train,predictions_train)))
print("Accuracy_test : "+str(accuracy_score(y_test,predictions_test)))
```

```python
y_prob = model_LR.predict_proba(X_test)[:,1] # This will give you positive class prediction probabilities  
y_pred = np.where(y_prob > 0.5, 1, 0) # This will threshold the probabilities to give class predictions.
model_LR.score(X_test, y_pred)
```

```python
confusion_matrix=metrics.confusion_matrix(y_train,predictions_train)
confusion_matrix

from sklearn.metrics import classification_report
print(classification_report(y_train,predictions_train))
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,predictions_test)
confusion_matrix

print(classification_report(y_train,predictions_train))
```

# K Neighbors Classifier

```python
# K Neighbors Classifier

model_KNN = KNeighborsClassifier()

# Run model on Train data
model_KNN.fit(X_train, y_train)

# Predictions
predictions_train = model_KNN.predict(X_train)
predictions_test = model_KNN.predict(X_test)

# Accuracy
from sklearn.metrics import accuracy_score
print("Accuracy_train : "+str(accuracy_score(y_train,predictions_train)))
print("Accuracy_test : "+str(accuracy_score(y_test,predictions_test)))
```

```python
confusion_matrix = metrics.confusion_matrix(y_train,predictions_train)
confusion_matrix

from sklearn.metrics import classification_report
print(classification_report(y_train,predictions_train))
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,predictions_test)
confusion_matrix

print(classification_report(y_train,predictions_train))
```
